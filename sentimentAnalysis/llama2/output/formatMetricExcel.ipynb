```python
import ast
```

```python
# Read the contents of the input file
with open('input.txt', 'r') as file:
    data = file.read()

# Convert the string representation of dictionary to a dictionary object
metrics = ast.literal_eval(data)
```

```python
# Define the order of keys according to the desired output
keys_order = [
    'eval_loss',
    'eval_accuracy',
    'eval_mae',
    'eval_precision',
    'eval_recall',
    'eval_f1',
    'eval_runtime',
    'eval_samples_per_second',
    'eval_steps_per_second',
    'epoch',
    'eval_class_0_accuracy',
    'eval_class_0_precision',
    'eval_class_0_recall',
    'eval_class_0_f1',
    'eval_class_1_accuracy',
    'eval_class_1_precision',
    'eval_class_1_recall',
    'eval_class_1_f1',
    'eval_class_2_accuracy',
    'eval_class_2_precision',
    'eval_class_2_recall',
    'eval_class_2_f1',
    'eval_class_3_accuracy',
    'eval_class_3_precision',
    'eval_class_3_recall',
    'eval_class_3_f1',
    'eval_class_4_accuracy',
    'eval_class_4_precision',
    'eval_class_4_recall',
    'eval_class_4_f1',
]
```

```python
# Create the desired output string
output = '\n'.join([f'="{metrics[key]}"' for key in keys_order])

# Write the output to a file
with open('output.txt', 'w') as file:
    file.write(output)
```
