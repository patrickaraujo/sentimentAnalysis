```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
# Load the pre-trained BERT model and tokenizer
#mod = "mistralai/Mistral-7B-v0.1"
model_name = "bert-base-uncased"
pretrained = "/content/drive/MyDrive/H.IAAC/datasets/bert_sentiment_model"
model = AutoModelForSequenceClassification.from_pretrained(pretrained)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/H.IAAC/datasets/filtered_output_rest.csv')
```

```python
dataset
```

```python
comments = dataset['content']
```

```python
predictions = []
i = 0
for comment in comments:
  encoded_input = tokenizer(comment, return_tensors="pt")
  output = model(**encoded_input)
  prediction = torch.argmax(output.logits, dim=-1)
  sentiment = "Negative" if prediction.item() == 0 else "Neutral" if prediction.item() == 1 else "Positive"
  prediction = prediction.item()
  predictions.append(prediction)
  i = i+1
  print(f'Comment {str(i)}: {comment}\tPrediction: {sentiment}')
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Assuming you have true labels (y_true) and predicted labels (y_pred)
accuracy = accuracy_score(dataset['sentiment'], predictions)
precision_macro = precision_score(dataset['sentiment'], predictions, average='macro')
recall = recall_score(dataset['sentiment'], predictions, average='macro')
f1 = f1_score(dataset['sentiment'], predictions, average='macro')
conf_matrix = confusion_matrix(dataset['sentiment'], predictions)

print("Accuracy:", accuracy)
print("Precision (Macro):", precision_macro)
print("Recall (Macro):", recall)
print("F1 Score (Macro):", f1)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
# Prepare text input
text = "Real good Good deals Get there fast Real quick and good"
encoded_input = tokenizer(text, return_tensors="pt")
```

```python
# Perform inference
output = model(**encoded_input)
pred = torch.argmax(output.logits, dim=-1)
```

```python
# Decode sentiment
sent = "Negative" if pred.item() == 0 else "Neutral" if pred.item() == 1 else "Positive"

print(f"Text: {text}")
print(f"Sentiment: {sent}")
```
